```python
def DataPrep(raw_data):
    Data = {}
    for index,row in raw_data.iterrows():
        if int(row['userId']) not in Data:
            Data[int(row['userId'])] = {}
        if(row['rating']>=3):
            Data[int(row['userId'])][int(row['movieId'])] = row['rating']
    return Data
```

```python
data = DataPrep(movie)
```

```python
def CompMatrix(train):
    C = {}
    N = {}
    for users,items in train.items():
#         print(users,items)
        for i in items:
            if i not in N:
                N[i] = 1
            else:
                N[i] += 1
            for j in items:
                if i==j:
                    continue
                if i not in C:
                    C[i] = {j:1}
                else:
                    if j not in C[i]:
                        C[i][j] = 1
                    else:
                        C[i][j] += 1
    return C,N
```

```python
C,N = CompMatrix(data)
```

```python
def CalItemSimilarity(C,N):
    W = {}
    for i, related_items in C.items():
        for j,Cij in related_items.items():
            temp = Cij/math.sqrt(N[i]*N[j])
            if i not in W:
                W[i] = {j:temp}
            if j not in W[i]:
                W[i][j] = temp
    return W
```

```python
Weight = CalItemSimilarity(C,N)
```

```python
def splitYvalue(data, seed,m,k):
    train_set = {}
    y_set = {}
    random.seed(seed)
    for user in data:
        y_set[user] = {}
        train_set[user] = {}
        for movie in data[user]:
            if random.randint(0,m)<k:
                y_set[user][movie] = data[user][movie]
            else:
                train_set[user][movie] = data[user][movie]
#         train = movies[:int(0.8*len(movies))]
#         y = movies[int(0.8*len(movies)):]
    #print('Data preparation finished')
    return train_set, y_set
```

```python
train, y_set = splitYvalue(data,2,10,2.5)
print(len(train[1]),len(y_set[1]))
```

```python
def Recommendation(train,user_id,W,k):
    rank = {}
    ru = train[user_id] #get the set of movies watched by user_id
    for i in ru:
        #iterate all movies j that user_id havent watched yet
        for j,wij in sorted(W[i].items(),key = itemgetter(1),reverse = True)[0:k]:
            if j in ru:
                continue
            #get the rating of movie i rated by user_id
            pi = ru[i]
            if j in rank:
                rank[j] += pi*wij
            else:
                rank[j] = pi*wij
    return rank
```
